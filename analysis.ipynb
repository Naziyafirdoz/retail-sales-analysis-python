### Feature Engineering

#### Q3. Create new columns

```python
df["Year"] = df["Date"].dt.year

df
```

```python
df["Month"] = df["Date"].dt.month

df
```

```python
df["Month Name"] = df["Date"].dt.month_name()

df
```

```python
df["Day Name"] = df["Date"].dt.day_name()

df
```

#### Q4. Business Columns

```python
def order_priority(x):
    if x > 600:
        return "High"
    elif x > 300:
        return "Medium"
    else:
        return "Low"

df["Order Priority"] = df["Total Amount"].apply(order_priority)

df
```

```python
def categorize_age(x):
    if x < 25:
        return "Young"
    elif x < 45:
        return "Adult"
    else:
        return "Senior"

df["Age Group"] = df["Age"].apply(categorize_age)

df
```

```python
df["Day Number"] = df["Date"].dt.weekday

df
```

```python
df["Day Type"] = df["Day Number"].apply(
    lambda x: "Weekend" if x > 4 else "Weekday"
)

df
```

### Core Analysis

#### Q5. Category Performance

```python
df.groupby("Product Category")["Total Amount"].sum().reset_index().sort_values("Total Amount",ascending = False)
```

##### Catergory Insight:

Electronics generated the highest revenue ($156,905), followed closely by Clothing. This indicates strong demand for electronic products. The business should focus on expanding electronics inventory and running targeted promotions.

#### Q6. Gender Analysis

```python
df.groupby("Gender")["Total Amount"].sum().reset_index().sort_values("Total Amount", ascending = False)
```

##### Gender Insight:

Female customers contributed slightly more revenue ($232,840) compared to male customers.This suggests higher purchasing activity among females. The company can introduce targeted marketing campaigns for female customers to increase sales further.

#### Q7. Time Analysis

```python
df.groupby("Month Name", observed = False)["Total Amount"].sum().reset_index().sort_values("Total Amount", ascending = False)
```

##### Monthly Insight:

May recorded the highest sales ($53,150), indicating peak seasonal demand. The business should increase stock levels and run promotional campaigns during this period to maximize revenue.

```python
df.groupby("Day Name")["Total Amount"].sum().reset_index().sort_values("Total Amount", ascending = False).head(1)
```

##### Weekly Insight:

Saturday generated the highest sales ($78,815), showing strong weekend shopping behavior. The company should introduce weekend discounts and offers to boost sales.

#### Q8. Customer Behavior

```python
df.groupby("Age Group")["Total Amount"].sum().reset_index().sort_values("Total Amount", ascending = False)
```

##### Customer Insight:

Adult customers contributed the highest revenue ($193,925), making them the most valuable customer segment. Marketing strategies should primarily target this age group.

```python
df[df["Total Amount"] > 500].groupby("Product Category")["Total Amount"].count().reset_index()
```

##### High-Value Orders Insight:

High-value transactions (>$500) are distributed across all categories, indicating demand for premium products. The company can introduce premium bundles across categories.

### Visualizations

```python
import matplotlib.pyplot as plt

df.pivot_table(
    values="Total Amount",
    index="Product Category",
    aggfunc="sum"
).plot(kind="bar")

plt.title("Revenue by Product Category", fontsize = 14)
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=0)
plt.show()
```

```python
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

df["Month Name"] = pd.Categorical(df["Month Name"], categories=month_order, ordered=True)
```

```python
df.pivot_table(
    values = "Total Amount",
    index = "Month Name",
    aggfunc = "sum",
    observed = False
).sort_index().plot(kind = "line")

plt.title("Monthly trends", fontsize = 14)
plt.xlabel("Month Name")
plt.ylabel("Total Revenue")
plt.show()
```

```python
df.pivot_table(
    values = "Total Amount",
    index = "Gender",
    aggfunc = "sum"
).plot(kind = "bar")

plt.title("Gender Comparison", fontsize = 14)
plt.xlabel("Gender")
plt.ylabel("Total Revenue")
plt.show()
```

### Business Insights

1. The Electronics category generates the highest revenue, indicating strong customer demand. The business should focus on expanding this category and increasing stock availability.

2. Female customers contribute slightly more to overall sales, suggesting higher engagement. Targeted promotions can further improve retention.

3. Sales peak during certain months (like May), indicating seasonal demand. The company should plan inventory and marketing campaigns accordingly.

4. Weekend sales are higher compared to weekdays, showing strong shopping behavior during weekends. Special weekend offers can boost revenue further.

5. Adult customers contribute the highest revenue, making them the most valuable segment. Marketing strategies should focus on this age group.

6. High-value transactions (above ₹500) are spread across all categories, indicating consistent demand for premium products. The company can introduce bundled offers or loyalty programs for high-spending customers.
